# pileUp filter for Run2 pp analysis


For run 2 pp analysis, the pileup scenario (pileup probability) for 5 TeV reference run is around 1.2-1.4, where the pileup is defined as <dfn><strong> multiple interactions per bunch crossing </dfn></strong>. The pileup events, if not excluded, would have a non-negligible effect to various of physics observables. For the simplest case, the pp spectra measurement, we want to measure the particle yield, or number of tracks in each $p_{T}$ bin, per event instead of per bunch crossing. Therefore, the pileup filter is necessary for precise measurements.

### <hr> Types of filters

Multiple collisions events can be identified by the number of good primary vertices in the event. However, not all multiple collision events would lead to multiple reconstructed vertex, nor single collision event would lead to only one reconstructed vertex. In the multiple collision events with only one vertex reconstructed, the two collisions could be close enough that vertex reconstruction algorithm fails. In the case of of a single collisions, the decay of a short lived particle close to the beam line may result in the reconstruction of an additional primary vertex (vertex splitting).

In order to balance between statistics and pileup contamination, one needs to look at the spatial distribution of the vertices, the correlation between number of tracks associated with the second primary vertex and dz value, to come up with a reasonable cut for the pileup filter. 

- <strong> dxy cut </strong>: accept those events that have two vertices where dxy is larger than some value, i.e, <strong> 500 $\mu m$ </strong>

- <strong> dz cut </strong>: accept those events that have two vertices where dz is larger than some value, i.e, <strong> 1 $cm$ </strong>

- <strong> Ntrk vs dz cut</strong>: according to the number of tracks w.r.t the second primary vertex, cut on a dz value

- <strong> tight cut</strong>: accept events that only has 1 vertex. 

<strong> Different types of filter can be obtained by combining different cuts. In HIN-15-006, HIN-15-009, FSQ-15-002, the pileup filter was,</strong>

dz (1cm) + dxy (500$\mu m$) + dzNtrkCut 

### Available filters now

- <strong>  pileupFilter_baseCut_loose </strong>, with dxy cut and dzNtrk loose cut (pink line)
- <strong>  pileupFilter_baseCut_tight </strong>, with dxy cut and dzNtrk tight cut (black line)
- <strong>  pileupFilter_baseCut_loose_dz1p0 </strong>, with dxy cut and dzNtrk loose cut (pink line) plus dz cut 1 cm
- <strong>  pileupFilter_baseCut_tight_dz1p0 </strong>, with dxy cut and dzNtrk tight cut (black line) plus dz cut 1 cm
- <strong>  pileupFilter_vtx1 </strong>, with only 1 reconstructed vertex
